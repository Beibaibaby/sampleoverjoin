```python
# scale disjoint
import matplotlib.pyplot as plt
import numpy as np

scales = [1, 3, 5, 10]

exact = [136.58558659098344, 188.71116442902712, 241.30457305302843, 357.3763971120352]

olken = [633.2847298400011, 1263.3742737659486, 1924.9753420220222, 3703.2599187419983]
# 159267, 155551, 150374, 159135

fig, ax = plt.subplots(figsize=[9, 7])

plt.plot(scales, exact,'o-', label='exact + exact',markersize=8)

plt.plot(scales, olken,'o-', label='direct + olken',markersize=8)

ax.set_xlabel('original TPCH scale factor (GB)')
ax.set_ylabel('time (seconds)')

plt.title('UQ1, disjoint union, 10^4 sample size')
# plt.yscale("log")
# plt.xscale("log")  
plt.legend()
plt.show() 
```
